Scrapping a news website & creating a dataframe.

```python
import requests
w = "https://www.politifact.com/factchecks/?page=1"
result = requests.get(w)
content = result.text

from bs4 import BeautifulSoup
soup = BeautifulSoup(content,'html.parser')
```

```python
st = soup.findAll('div',class_="m-statement__quote")
s = []
l = []
for i in st:
    a = i.find('a',href=True).text.strip()
    s.append(a)
    b = 'https://www.politifact.com'+i.find('a',href=True)['href']
    l.append(b)
s
```

```python
dc = soup.findAll('footer',class_='m-statement__footer')
dc1 = []
for i in dc:
    q = i.text.strip().split('•')[1]
    dc1.append(q)
dc1
```

```python
src = soup.findAll('a',class_='m-statement__name',href=True)
src1 = []
for i in src:
    src1.append(i.text.strip())
src1
```

```python
p = soup.findAll('div',class_='m-statement__meter')
p1 = []
for i in p:
    p1.append(i.find('img',class_='c-image__original')['alt'])
p1
```

```python
import pandas as pd
data = pd.DataFrame({'News':s,'Date':dc1,'Source':src1,'Meter':p1})
```

```python
import requests
from bs4 import BeautifulSoup
# Send a GET request to the website
url = "https://indianexpress.com/"
response = requests.get(url)
# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, "html.parser")
# Find the div with class "top news"
top_news_div = soup.find("div", {"class":"top-news"})
# Find all the headline links within the top news div
headline_links = top_news_div.findAll("a")
# Extract and print the headlines
for link in headline_links:
    headline = link.text.strip()
    print(headline)
```

```python
for i in top_news_div:
    headline_links = i.findAll("a",href=True)
    print(headline_links[0].text.strip()) 
```

```python
news_headlines = soup.find("div",{"class":"right-part bg"})
print(news_headlines.prettify())
```

```python
links=news_headlines.find("div",{"class":"top-news"})
for i in links.findAll("li"):
    print(i.find("a").text.strip())
```
