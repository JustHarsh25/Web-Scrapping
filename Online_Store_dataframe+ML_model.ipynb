#Creating a dataframe from online store website & creating Linear Regression model to predict actual price

```python
from bs4 import BeautifulSoup
f = open('All Categories- Buy Products Online at Best Price in India - All Categories _ Flipkart.com.html','r',encoding='utf-8')
soup =BeautifulSoup(f,'html.parser')
x = soup.findAll('div',class_='_4rR01T')
l1 = []
for i in x:
    l1.append(i.text.strip())
print(len(l1))
```

```python
y = soup.find_all('li',class_='rgWa7D')
l2 = []
for i in range(0,len(y),6):
    l2.append(y[i].text.strip())
len(l2)
```

```python
z = soup.find_all('div',class_='_1_WHN1')
l3 = []
for i in z:
    l3.append(int(i.text.strip().split('₹')[1].replace(',','')))
l3
```

```python
w = soup.find_all('div',class_='_3I9_wc')
l4 = []
for i in w:
    l4.append(int(i.text.strip().split('₹')[1].replace(',','')))
l4
```

```python
import pandas as pd
data = pd.DataFrame({'Mobile_Name':l1,'Specs':l2,'Selling_price':l3,'Actual_price':l4})
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
x_train,x_test,y_train,y_test = train_test_split(data[['Actual_price']],data['Selling_price'],test_size=0.2,random_state=2)
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
print(r2_score(y_test,y_pred))
```
