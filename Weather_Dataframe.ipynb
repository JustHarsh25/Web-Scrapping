```python
import pandas as pd
url2=f'http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}'
response = requests.get(url2)
res2 = response.json()
x = (json.dumps(res2,indent=4))
d_t = []
temp = []
pr = []
hu = []
wd = []
ws = []
for i in range(len(res2['list'])):
    temp.append(res2['list'][i]['main']['temp'])
    d_t.append(res2['list'][i]['dt_txt'])
    pr.append(res2['list'][i]['main']['pressure'])
    hu.append(res2['list'][i]['main']['humidity'])
    wd.append(res2['list'][i]['weather'][0]['description'])
    ws.append(res2['list'][i]['wind']['speed'])
data = pd.DataFrame({'date_time':d_t,'temp':temp,'pressure':pr,'humidity':hu,'weather_description':wd,'wind_speed':ws})
```
